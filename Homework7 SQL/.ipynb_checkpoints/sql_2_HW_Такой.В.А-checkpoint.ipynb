### Все задания нужно выполнить только с помощью sql

1. Выведите количество пустых значений по колонкам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во пустых значений)

```python
pd.read_sql(
    '''
    SELECT 
        'CRIM' AS column_name, 
        SUM(CASE WHEN CRIM IS NULL OR CRIM = 0 THEN 1 ELSE 0 END) AS missing_values
    FROM boston
    UNION ALL
    SELECT 
        'ZN', 
        SUM(CASE WHEN ZN IS NULL OR ZN = 0 THEN 1 ELSE 0 END)
    FROM boston
    UNION ALL
    SELECT 
        'INDUS', 
        SUM(CASE WHEN INDUS IS NULL OR INDUS = 0 THEN 1 ELSE 0 END)
    FROM boston
    UNION ALL
    SELECT 
        'CHAS', 
        SUM(CASE WHEN CHAS IS NULL OR CHAS = 0 THEN 1 ELSE 0 END)
    FROM boston
    UNION ALL
    SELECT 
        'NOX', 
        SUM(CASE WHEN NOX IS NULL OR NOX = 0 THEN 1 ELSE 0 END)
    FROM boston;
    ''',
    con
)
```

2. Выведите количество уникальных значений по колонокам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во уникальных значений)

```python
pd.read_sql(
    '''
    SELECT
        'CRIM' as column, COUNT(DISTINCT CRIM) as unique_count
    FROM boston
    UNION ALL
    SELECT
        'ZN', COUNT(DISTINCT ZN)
    FROM boston
    UNION ALL
    SELECT
        'INDUS', COUNT(DISTINCT INDUS)
    FROM boston
    UNION ALL
    SELECT
        'CHAS', COUNT(DISTINCT CHAS)
    FROM boston
    UNION ALL
    SELECT
        'NOX', COUNT(DISTINCT NOX)
    FROM boston
    ''',
    con
)
```

3. Выведите колонки, у которых медиана равна минимальному значению (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX.
Напишите какой вывод можно сделать по данным в этих колонках

```python
pd.read_sql(
    '''
    SELECT column_name
    FROM (
        SELECT 'CRIM' as column_name, MIN(CRIM) as min_val, 
               (SELECT CRIM FROM boston ORDER BY CRIM LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston)/2) as median_val
        FROM boston
        UNION ALL
        SELECT 'ZN', MIN(ZN), 
               (SELECT ZN FROM boston ORDER BY ZN LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston)/2)
        FROM boston
        UNION ALL
        SELECT 'INDUS', MIN(INDUS), 
               (SELECT INDUS FROM boston ORDER BY INDUS LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston)/2)
        FROM boston
        UNION ALL
        SELECT 'CHAS', MIN(CHAS), 
               (SELECT CHAS FROM boston ORDER BY CHAS LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston)/2)
        FROM boston
        UNION ALL
        SELECT 'NOX', MIN(NOX), 
               (SELECT NOX FROM boston ORDER BY NOX LIMIT 1 OFFSET (SELECT COUNT(*) FROM boston)/2)
        FROM boston
    )
    WHERE min_val = median_val
    ''',
    con
)
```

Запрос выявил, что медиана и минимальное значение совпадают в двух колонках: ZN и CHAS.

Вывод:

ZN (доля земли, отведенной под участки более 25 000 кв. футов): Медиана этой переменной равна нулю, что означает, что более половины районов не имеют участков, отведенных под большие дома.
CHAS (граница реки Чарльз): Медиана также равна нулю, что указывает на то, что большинство домов не находятся на границе реки Чарльз.
Таким образом, большинство районов Бостона не включает крупные участки земли и не граничит с рекой Чарльз.""

4. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг) среди всех значений LSTAT. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(
    '''
    WITH ranks AS (
        SELECT 
            LSTAT, 
            MEDV,
            RANK() OVER (ORDER BY LSTAT) as rank
        FROM boston
    )
    SELECT LSTAT, MEDV, rank
    FROM ranks
    WHERE MEDV = (SELECT MAX(MEDV) FROM boston)
    ''',
    con
)
```

Запрос показал, что самые дорогие дома имеют низкие значения LSTAT (процент населения с более низким статусом). Эти значения варьируются от 1.73 до 9.53, и все они находятся на низких рангах в выборке (от 1 до 203).

Вывод:

Более дорогие дома, как правило, находятся в районах с меньшим процентом населения с низким социальным статусом, что может свидетельствовать о социальной и экономической сегрегации.

5. Выведите среднюю стоимость домов граничащих с рекой(CHAS) и нет (граничит/не граничит, стоимость)

```python
pd.read_sql(
    '''
    SELECT 
        CHAS as "Граничит с рекой", 
        AVG(MEDV) as "Средняя стоимость"
    FROM boston
    GROUP BY CHAS
    ''',
    con
)
```

6. Выведите все колонки, у которых среднее значение выше, когда дом граничит с рекой (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(
    '''
    WITH MedianValues AS (
    SELECT 
        CRIM, ZN, INDUS, CHAS, NOX,
        ROW_NUMBER() OVER (ORDER BY CRIM) as crim_row,
        ROW_NUMBER() OVER (ORDER BY ZN) as zn_row,
        ROW_NUMBER() OVER (ORDER BY INDUS) as indus_row,
        ROW_NUMBER() OVER (ORDER BY CHAS) as chas_row,
        ROW_NUMBER() OVER (ORDER BY NOX) as nox_row,
        (SELECT COUNT(*) FROM boston) as row_count
    FROM boston
),
MedianComparison AS (
    SELECT
        'CRIM' as column_name, 
        (SELECT MIN(CRIM) FROM boston) as min_val, 
        (SELECT CRIM FROM MedianValues WHERE crim_row = (row_count + 1) / 2) as median_val
    UNION ALL
    SELECT 
        'ZN', 
        (SELECT MIN(ZN) FROM boston), 
        (SELECT ZN FROM MedianValues WHERE zn_row = (row_count + 1) / 2)
    UNION ALL
    SELECT 
        'INDUS', 
        (SELECT MIN(INDUS) FROM boston), 
        (SELECT INDUS FROM MedianValues WHERE indus_row = (row_count + 1) / 2)
    UNION ALL
    SELECT 
        'CHAS', 
        (SELECT MIN(CHAS) FROM boston), 
        (SELECT CHAS FROM MedianValues WHERE chas_row = (row_count + 1) / 2)
    UNION ALL
    SELECT 
        'NOX', 
        (SELECT MIN(NOX) FROM boston), 
        (SELECT NOX FROM MedianValues WHERE nox_row = (row_count + 1) / 2)
)
SELECT column_name
FROM MedianComparison
WHERE min_val = median_val;

    ''',
    con
)
```

Запрос выявил, что колонки CRIM, ZN, INDUS, NOX имеют более высокие средние значения, когда дом граничит с рекой.

Вывод:

CRIM (уровень преступности): Уровень преступности выше в районах, расположенных рядом с рекой, что может указывать на то, что река может служить фактором, влияющим на преступность.
ZN, INDUS, NOX: Также более высокая доля земли, отведенной под жилую застройку, более высокая доля промышленных зон и более высокая концентрация оксидов азота наблюдаются в этих районах.

7. Выведите значения долей промышленной застройки(INDUS), концентрации оксидов азота(NOX) и по их перцентилям - 10, 20 ... 100 ( перцетиль(10,20...100),значение INDUS, значение NOX). Напишите прослеживается между ними взаимосвязь

```python
pd.read_sql(
    '''
    WITH percentiles AS (
        SELECT
            PERCENT_RANK() OVER (ORDER BY INDUS) AS percentile_rank_indus,
            INDUS,
            PERCENT_RANK() OVER (ORDER BY NOX) AS percentile_rank_nox,
            NOX
        FROM boston
    )
    SELECT 
        ROUND(percentile_rank_indus * 10) * 10 AS indus_percentile,
        MIN(INDUS) AS indus_value,
        ROUND(percentile_rank_nox * 10) * 10 AS nox_percentile,
        MIN(NOX) AS nox_value
    FROM percentiles
    GROUP BY indus_percentile, nox_percentile
    ORDER BY indus_percentile, nox_percentile;
    ''',
    con
)
```

"Запрос показывает, что значения долей промышленной застройки (INDUS) и концентрации оксидов азота (NOX) коррелируют между собой. По мере увеличения перцентиля INDUS наблюдается увеличение NOX.
Вывод:
Между индустриализацией района (INDUS) и загрязнением воздуха (NOX) существует прямая взаимосвязь: чем выше доля промышленной застройки, тем выше концентрация оксидов азота в воздухе."
